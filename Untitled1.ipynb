````python
import numpy
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
from time import time
````

````python
def betaPDF(x, a, b):
    betaPDF = (math.gamma(a+b)/(math.gamma(a)*math.gamma(b)))*(x**(a-1))*((1-x)**(b-1))
    return betaPDF
def uniformPDF(x, a, b):
    if x>=a and x<=b:
        return 1/(b-a)
    else:
        return 0
def WignerCDF(x, R):
    F=(1/(4*R*R*R))*(2*R-x)*(R+x)*(R+x)
    return F
def WignerPDF(x, R):
    f=(3/(4*R*R*R))*(R*R-x*x)
    return f
````

#Inverse Smapling via Newtons Method (2)

````python

def Newton_RaphsonSecond(r,u):
    loopBoolean= True
    x=numpy.random.uniform(low=-r,high=r) # intialize x in correct boundary
    prevX=0
    i=0

    while loopBoolean:
        prevX=x
        i+=1
        x=x-(WignerCDF(x,r)-u)/WignerPDF(x,r) # main part where we update our x

        if abs(prevX-x)<0.00001: # our tolerance
            loopBoolean=False
        if i>500:
            print("Too many iterations, something might be wrong")
            break
    return x


samples=10000
i=0
x=list()
r=1
t0=time()
while i<samples:
    tempX=10
    while abs(tempX)>1:# Since f(x) is bounded by -1 and 1 we dont want to go beyond that
        tempU=numpy.random.uniform(low=0,high=1) # Generate a random number from the uniform distribution (0,1)
        tempX=Newton_RaphsonSecond(r,tempU)# Find x such that, 0=F(x)-u
    x.append(tempX)
    i+=1
t1=time()
print("Newton method took "+ str((t1-t0))+" Seconds")
plt.hist(x,bins=50, density=True)
plt.xlabel("Random Variable X")
plt.ylabel("Frequency")
plt.title("Inverse Sampling via Newtons Method")
plt.show()
````

# Inverse Sampling via Bisection Method

````python


def Bisection_Method(r,u):

    # find proper a such that F(a)<0
    findA= False
    a=0
    while not findA:
        a=numpy.random.uniform(low=-r,high=r) # Generate number that is in bounds of our cdf
        if (WignerCDF(a,r)-u)<0:
            findA=True
            a=a

    # find proper B such that F(b)>0
    b=0
    findB= False
    while not findB:
        b=numpy.random.uniform(low=-r,high=r) # Generate number that is in bounds of our cdf
        if (WignerCDF(b,r)-u)>0:
            findB=True
            b=b

    # now find the roots using binary Search
    loopBoolean= True
    i=0
    while loopBoolean:

        c=(a+b)/2
        if (WignerCDF(c,r)-u)<0:
            a=c
        elif (WignerCDF(c,r)-u)>0:
            b=c
        if abs(b-a)<0.01 or abs(WignerCDF(c,r)-u)<0.0001:
            loopBoolean=False
        i+=1
    return c

samples=10000
i=0
x=list()

t0=time()
while i<samples:
    tempX=10
    while abs(tempX)>1: # Since f(x) is bounded by -1 and 1 we dont want to go beyond that
        tempU=numpy.random.uniform(low=0,high=1)
        tempX=Bisection_Method(1,tempU)
    x.append(tempX)
    i+=1
t1=time()
print("Bisection method took "+ str((t1-t0))+" Seconds")

plt.hist(x,bins=50,density=True)
plt.xlabel("Random Variable X")
plt.ylabel("Frequency")
plt.title("Inverse Smapling Via Bisection Method")
plt.show()
````

The Newton method typically takes between 0.15 to 0.46 seconds on average to generate 10,000 samples, whereas the Bisection method averages between 0.32 to 0.66 seconds for the same task...<br>
Newton's method exhibits quadratic convergence when the initial x is close to the root, whereas the Bisection method boasts a runtime of O(log(n))...<br>
Under suitable circumstances, the Newton method can outpace the Bisection method...<br>
Potential Pitfalls for the Newton method include:<br>
* Longer runtime if x is not initialized near the root.<br>
* Possibility of non-convergence if local minima or maxima exist between the initialized x and the root...<br>

Potential Pitfalls for the Bisection method include:<br>
* Inability to use the algorithm if the root merely touches the x-axis without crossing it; this is because we require points 'a' and 'b' such that f(a) < 0 < f(b)

I will make some graph to visualize and understand Wigner Semi-circle Distribution

````python

# Code to visualize PDF Of Wigner Semi-Distribution
domainWigner=numpy.linspace(-1.1,1.1,100)
R=1
WignerDensity=list()
for i in domainWigner:
  WignerDensity.append(WignerPDF(i,R))
plt.plot(domainWigner,WignerDensity)
plt.ylim(0,1)
plt.title("Wigner Semi-Circle Distribution")
plt.show()

# Code to Visualize the CDF of Wigner Semi-circle Distribution
y=list()
for i in domainWigner:
  y.append(WignerCDF(i,R))
plt.plot(domainWigner,y)
plt.ylim(0,1.2)
plt.title("Wigner Semi-Circle Distribution CDF")
plt.show()

# code to visualize beta(1,1)
y=list()
for i in domainWigner:
  y.append(betaPDF(i,1,1))
plt.plot(domainWigner,y)
plt.ylim(0,1.2)
plt.title("Beat(1,1) PDF")
plt.show()

# code to visualize beta(1/2,1/2)
y=list()
for i in domainWigner:
  y.append(betaPDF(i,1/2,1/2))
plt.plot(domainWigner,y)
plt.ylim(0,3)
plt.title("Beta(1/2, 1/2) PDF")
plt.show()

# code to visualize beta(3/2,3/2)
y=list()
for i in domainWigner:
  y.append(betaPDF(i,3/2,3/2))
plt.plot(domainWigner,y)
plt.ylim(0,1.5)
plt.title("Beta(3/2,3/2) PDF")
plt.show()


````

#(1) Using beta(1,1) to sample the the Wigner Distribution


````python
alpha = 1
beta_val = 1
samples = 10000
i = 0
x = list()
c=1.5
while i < samples:
    X=stats.beta.rvs(alpha, beta_val,-1,2)
    # X=numpy.random.beta(alpha,beta_val)
    U=numpy.random.uniform(low=0,high=1)
    # stats.beta.pdf(X,alpha,beta_val,-1,2)
    if U<=WignerPDF(X,1)/(c*stats.beta.pdf(X,alpha,beta_val,-1,2)):
        x.append(X)
        i+=1
plt.hist(x, bins=80,density=True, label="Resulting density from Rejection Algorithm")
plt.plot(domainWigner, WignerDensity, label="Pdf of Wigner Semi-Circler")
plt.legend()
plt.ylim(0,1)
plt.title("Rejection algortihm using Beta(1,1)")
plt.show()
````
